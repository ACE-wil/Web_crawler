```python

session = HTMLSession()

r = session.get(url,headers=wuyou_job_headers)

response_data = r.json()

response_data
```

```python
df_wuyou = pd.json_normalize(response_data['resultbody']['job']['items'])
df_wuyou
```

```python
citi_name = str(城市)
citi_name
```

```python
key = str(关键字)
key
```

```python
import time
time.localtime()
output_time = str(time.localtime().tm_year)+'-'\
              +str(time.localtime().tm_mon)+'-'\
             +str(time.localtime().tm_mday)+'_'\
             +str(time.localtime().tm_hour)+'-' \
             +str(time.localtime().tm_min)
output_time
```

```python
df_wuyou.to_excel('前程无忧'+citi_name+key+output_time+'.xlsx')
```
