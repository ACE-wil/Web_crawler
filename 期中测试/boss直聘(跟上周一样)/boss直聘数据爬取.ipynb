```python
cities={
    '北京':'101010100',
    '上海':'101020100',
    '广州':'101280100',
    '深圳':'101280600'
}
```

```python
def get_cities(city):
    return cities[str(city)]
```

```python
关键词 = input("输入职位关键词")
```

```python
城市 = input("输入城市，目前支持北京、上海、广州、深圳")
```

```python
import urllib.parse
encoded_content = urllib.parse.quote(str(关键词))
print(encoded_content)
```

```python
from requests_html import AsyncHTMLSession
asession = AsyncHTMLSession()
r = await asession.get('https://www.zhipin.com/wapi/zpgeek/search/joblist.json?scene=1&query='+str(encoded_content)+'&city='+str(get_cities(城市))+'&experience=&payType=&partTime=&degree=&industry=&scale=&stage=&position=&jobType=&salary=&multiBusinessDistrict=&multiSubway=&page=1&pageSize=30',headers=headers)
```

```python
import time
import random
wait_time = random.randint(90,120)
```

```python
response_concat = []
for i in range(page):
    # import time
    # timestamp = time.time()
    # timestamp = int(timestamp * 1000)
    # token = session.get('https://www.zhipin.com/wapi/zppassport/get/zpToken?v='+str(timestamp),headers=headers2)
    # respon_token = token.json()
    # respon_token = respon_token['zpData']['token']
    # headers['Zp_token'] = respon_token
    time.sleep(wait_time)
    import pandas as pd
    from requests_html2 import HTMLSession
    session = HTMLSession()
    print('目前爬取到第'+str(i+1)+'页')
    from requests_html import AsyncHTMLSession
    asession = AsyncHTMLSession()
    r = await asession.get('https://www.zhipin.com/wapi/zpgeek/search/joblist.json?scene=1&query='+str(encoded_content)+'&city='+str(get_cities(城市))+'&experience=&payType=&partTime=&degree=&industry=&scale=&stage=&position=&jobType=&salary=&multiBusinessDistrict=&multiSubway=&page='+str(i+1)+'&pageSize=30',headers=headers)
    response_data = r.json()
    df_A = pd.json_normalize(response_data['zpData']['jobList'])
    response_concat.append(df_A)

df_boss = pd.concat(response_concat)
```

# 需要执行两年半，两年半后出来

```python
岗位要求 = []
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time
import random
# 设置 Chrome 为无头模式


for i in range(len(岗位详情页URL)):
    chrome_options = Options()
    chrome_options.add_argument("--headless")
    user_agent = "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    chrome_options.add_argument(f"user-agent={user_agent}")
# 初始化 Chrome WebDriver
    driver = webdriver.Chrome(options=chrome_options)
    url = str(岗位详情页URL[i])  # 替换为你要访问的网页 URL
    driver.get(url)
    # 等待一定时间，确保 JavaScript 代码有足够的时间加载数据
    # wait_time = random.randint(3)
    time.sleep(5)  # 你可能需要根据网页加载时间进行调整
    page_content = driver.page_source
    driver.quit()
    from bs4 import BeautifulSoup
    soup = BeautifulSoup(page_content, 'html.parser')
    command = soup.select_one('div.job-sec-text')
    岗位要求.append(command.text)
```
