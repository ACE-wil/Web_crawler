```python
import pandas as pd
```

```python
df = pd.read_excel(f'excels/猎聘深圳-广告策划-2023-12-23.xlsx')
```

```python
dates = []
for i in range(len(df['职位更新时间'])):
    years = str(df['职位更新时间'][i])[0:4]
    months = str(df['职位更新时间'][i])[4:6]
    days = str(df['职位更新时间'][i])[6:8]
    dates.append(years+'年'+months+'月'+days+'日')
dates
```

```python
df['职位更新时间'] = dates
```

```python
df
```

```python
for i in range(len(df['是否top职位'])):
    if df['是否top职位'][i] == False:
        df['是否top职位'][i] = '否'
    else:
        df['是否top职位'][i] = '是'
```

```python
df = df.drop('Unnamed: 0', axis=1)
```
