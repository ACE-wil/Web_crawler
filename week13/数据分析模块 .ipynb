# 分析recruiter.recruiterTitle

```python
df_ana['recruiter.recruiterTitle']
```

```python
df_ana['recruiter.recruiterTitle'].value_counts()
```

```python
data = [(k,v) for k,v in df_ana['recruiter.recruiterTitle'].value_counts().items()]
data
```

```python
import pyecharts.options as opts
from pyecharts.charts import WordCloud

c = (
    WordCloud()
    .add(series_name="热点分析", data_pair=data, word_size_range=[6, 66])
    .set_global_opts(
        title_opts=opts.TitleOpts(
            title="热点分析", title_textstyle_opts=opts.TextStyleOpts(font_size=23)
        ),
        tooltip_opts=opts.TooltipOpts(is_show=True),
    )
)
c.render_notebook()
```

# 分析Job


```python
df_ana.columns
```

```python
df_job = df_ana[['job.dq','job.title','job.labels']]
df_job
```

```python
list(set(list(df_job['job.dq'])))
```

```python
df_job[df_job['job.dq'].isin(['北京-石景山区'])]['job.title'].tolist()
```

```python
dict_区_岗位 = {i:df_job[df_job['job.dq'].isin([i])]['job.title'].tolist() for i in list(set(list(df_job['job.dq'])))}
dict_区_岗位
```

```python
['name','children']
```

```python
{'children':[
    { 'name':i ,'children':[{'name':j} for j in dict_区_岗位[i]]}   for i in list(set(list(df_job['job.dq'])))
],
 'name':'北京产品经理应届生'
}
```

```python
from pyecharts import options as opts
from pyecharts.charts import Tree

data = [
   {'children':[
    { 'name':"北京-丰台区" ,'children':[{'name':j} for j in dict_区_岗位['北京-丰台区']]} 
],
 'name':'北京产品经理应届生'
}]

c = (
    Tree()
    .add("", data, collapse_interval=2, layout="radial")
    .set_global_opts(title_opts=opts.TitleOpts(title="Tree-Layout"))

)
c.render_notebook()
```

```python




c = (
    Tree()
    .add("", data)
    .set_global_opts(title_opts=opts.TitleOpts(title="Tree-基本示例"))
)
c.render_notebook()
```

# 分析job.title and salary

```python
df_ana.columns
```

```python
df_ana[['comp.compName','job.title','job.salary']]
```

```python
df_ana['comp.compName'][0]
```

```python
from pyecharts.components import Table
from pyecharts.options import ComponentTitleOpts


table = Table()

headers = ['comp.compName','job.title','job.salary']
rows = []
for i in range(len(df_ana)):
    rows.append([df_ana['comp.compName'][i],df_ana['job.title'][i],df_ana['job.salary'][i]])
table.add(headers, rows)
table.set_global_opts(
    title_opts=ComponentTitleOpts(title="Table-基本示例", subtitle="我是副标题支持换行哦")
)
table.render_notebook()
```
