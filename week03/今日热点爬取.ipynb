```python
from requests_html import HTMLSession
session = HTMLSession()
```

```python
s = session.get('https://tophub.today/n/yjvQDpjobg')
```

```python
for i in range(len(s.html.find('td.al > a'))):
    product_title = s.html.find('td.al > a')
```

```python
for i in range(len(product_title)):
    print(product_title[i].text)
```

```python
output_time = str(time.localtime().tm_year)+'-'\
              +str(time.localtime().tm_mon)+'-'\
                +str(time.localtime().tm_mday)+'_'\
                 +str(time.localtime().tm_hour)+'-'\
                 +str(time.localtime().tm_min)
output_time
```

```python
import pandas as pd

taobao_product = pd.DataFrame()

for i  in range (len(s.html.find('td.al > a'))):
    taobao_product.loc[i,'商品标题'] = product_title[i].text.split(' ')[0]
    taobao_product.loc[i,'价格'] = product_title[i].text.split(' ')[1]
    taobao_product.loc[i,'链接'] = product_title[i].absolute_links
    taobao_product.to_excel('淘宝特价区'+output_time+".xlsx",index=False)
```

```python
# 下载图片
import requests
# 连接网页
p = session.get("https://tophub.today/n/yjvQDpjobg")
# 找图片属性
get_pic = p.html.find('td.al > img')
# 图片命名特殊字符
especail = ['+',"-","*",'/']


# 遍历所有img元素，返回索引和元素的元组序列
for i, img in enumerate(get_pic):
    img_url = img.attrs['src']
    img_name = str(product_title[i].text.split(' ')[0]) + '.jpg'

    # 检查图片名称中是否包含特殊字符 检查3遍
    for char in especail:
        if char in img_name:
            img_name = img_name.split(char)[0] + '.jpg'
            break
    for char in especail:
        if char in img_name:
            img_name = img_name.split(char)[0] + '.jpg'
            break
    for char in especail:
        if char in img_name:
            img_name = img_name.split(char)[0] + '.jpg'
            break
# 请求图片
    img_response = requests.get(img_url)
    # 把图片内容写进图片里
    with open('img/' + img_name, 'wb') as file:
        file.write(img_response.content)
```
